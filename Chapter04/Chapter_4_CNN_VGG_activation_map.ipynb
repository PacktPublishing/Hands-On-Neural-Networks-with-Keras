```python
from matplotlib import pyplot as plt
%matplotlib inline

from keras import activations
from keras.applications import VGG16
from vis.utils import utils
from vis.visualization import visualize_activation
```

```python
# Build the VGG16 network with ImageNet weights
model = VGG16(weights='imagenet', include_top=True)

# Utility to search for layer index by name. 
# Alternatively we can specify this as -1 since it corresponds to the last layer.
layer_idx = utils.find_layer_idx(model, 'predictions')

# Swap softmax with linear
model.layers[layer_idx].activation = activations.linear
model = utils.apply_modifications(model)
```

```python
# Define figure size
plt.rcParams['figure.figsize'] = (18, 6)

# 20 is the imagenet category for 'ouzel'
img = visualize_activation(model, layer_idx, filter_indices=288)
plt.imshow(img)
```

```python
img = visualize_activation(model, layer_idx, filter_indices=288, max_iter=500, verbose=True)
```

```python
plt.imshow(img)
```

```python
from vis.input_modifiers import Jitter

# 20 is the imagenet category for 'ouzel'
# Jitter 16 pixels along all dimensions to during the optimization process.
img = visualize_activation(model, layer_idx, filter_indices=288, max_iter=500, input_modifiers=[Jitter(16)])
plt.imshow(img)
```
